```python
def usecase_identification(prompt_):

  response = openai.Completion.create(
    model="davinci:ft-personal-2023-02-03-11-21-26",
    prompt=prompt_,
    max_tokens=30,
    temperature=0,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0,
    stop=["<end>"]
  )

  return response['choices'][0]['text']
```
